```python
def timeseries_split(X,y,test_size=0.1):
    index = X.shape[0] - round(X.shape[0]*test_size)
    X_train = X[:index]
    y_train = y[:index]
    X_test = X[index:]
    y_test = y[index:]
    return X_train, y_train, X_test, y_test

def prediction(X_train, test_index, regressor):
    y_pred = []
    sample = np.asarray(X_train.iloc[-1]).reshape(1, -1)
    for i in range(len(test_index)):
        pred = regressor.predict(sample)
        y_pred.append(pred[0])
        sample= np.append(sample[0][1:],pred).reshape(1, -1)
    y_pred = pd.Series(y_pred,index=test_index)
    return y_pred

def make_data_knn(data_w, L, cap_s):
    columns = [f'c{i}' for i in range(1,L+1)]+['t']
    lst=[np.asarray(data_w[cap_s][data_w.shape[0]-L-1-i:data_w.shape[0]-i]) for i in range(data_w.shape[0]-L)]
    lst.reverse()
    data_knn = pd.DataFrame(lst,columns=columns)
    y_knn = data_knn['t']
    data_knn = data_knn.drop(axis=1,columns='t')
    return data_knn, y_knn
    
```

```python
%%time
CaptureSites = list({f'CaptureSite_{i}' for i in range(29)} - {'CaptureSite_3'})
CaptureSites_dict=dict()
for site in CaptureSites:
    print(site)
    MRSE=999
    for window_size in range(2,5):
        data_knn, y_knn = make_data_knn(data_w, window_size, site)
        for n_neighbrs in range(2,5):
            NNregressor = KNeighborsRegressor(n_neighbors=n_neighbrs, p=2, weights='distance')
            tscv = TimeSeriesSplit(n_splits=3)
            mrse=np.array([])
            for train_index, test_index in tscv.split(data_knn):
                d_train= data_knn.iloc[train_index]
                y_train= y_knn.iloc[train_index]
                d_test=data_knn.iloc[test_index]
                y_test=y_knn.iloc[test_index]
                
                NNregressor.fit(d_train, y_train)
                y_pred = prediction(d_train, test_index, NNregressor)
                
                mrse=np.append(mrse,mean_squared_error(y_test, y_pred))
            if mrse.mean() < MRSE:
                MRSE=mrse.mean()
                CaptureSites_dict[site]=[site, window_size, n_neighbrs, mrse, mrse.mean()]
best_params = pd.DataFrame(list(CaptureSites_dict.values()),columns=['site','win_size','n_neigh','mrse','mrse_mean'])
```
