```python
lst=[]
CaptureSites = [f'CaptureSite_{i}' for i in range(29)]
month = [f'month_{i}' for i in range(1,13)]
cols = ['Year','N_week','count'] + CaptureSites + month
for y in [2010,2011,2012,2013,2014,2015,2016,2017,2018]:
    n_weeks = max(data[data.Year==y]['Week_num'])
    for week in range(1,n_weeks+1):
        row = []
        row.append(y)
        row.append(week)
        temp = data[(data.Week_num == week) & (data.Year == y)]
        row.append(temp.shape[0])
        for site in CaptureSites:
            row.append(temp[temp.CaptureSite == site].shape[0])
        for m in range(1,13):
            if m in temp.Month.unique():
                row.append(1)
            else:
                row.append(0)
        lst.append(row)

data_w = pd.DataFrame(lst,columns=cols)   
```

```python
# у 2010 и еще пары годов, вместо 52 недель - 53. 53 неделю удаляем.
data_w = data_w.drop(data_w[data_w.N_week == 53].index)
data_w.reset_index(drop=True,inplace=True)
```

```python
raw_data = data_w.loc[:,'count']
```

```python
def make_trajectory_matrix(data, L):
    columns = [f'c{i}' for i in range(1,L+1)]+['t']
    row_shape = data.shape[0]
    lst=[np.asarray(data[row_shape-L-1-i : row_shape-i]) for i in range(row_shape-L)]
    lst.reverse()
    data_= pd.DataFrame(lst,columns=columns)
    y_ = data_['t']
    data_ = data_.drop(axis=1,columns='t')
    return data_, y_
```

```python
data_for_training, target = make_trajectory_matrix(raw_data, 12)
```

```python
data_train = data_for_training.iloc[:-100,:]
y_train = target[:-100]

data_test = data_for_training.iloc[-100:,:]
y_test = target[-100:]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from IPython.display import clear_output
```

```python
data_train = torch.tensor(data_train.values, dtype=torch.float)
y_train = torch.tensor(y_train.values, dtype=torch.float)
y_train = y_train.unsqueeze(1)

data_test = torch.tensor(data_test.values, dtype=torch.float)
y_test = torch.tensor(y_test.values, dtype=torch.float)
y_test = y_test.unsqueeze(1)
```

```python
class TORT(nn.Module):
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
        nn.Linear(in_features=12, out_features=8, bias=True),
        nn.Tanh(),
        nn.Linear(in_features=8, out_features=4, bias=True),
        nn.Tanh(),
        nn.Linear(in_features=4, out_features=1, bias=True),
        )
    def forward(self,x):
        return self.layers(x)
```

```python
def train(model, optimizer, epochs = 800, epochs_for_paint = 50):
    train_loss=[]
    test_loss=[]
    loss_func=nn.MSELoss(reduction='mean')
    for epoch in range(0,epochs):
        optimizer.zero_grad()
        y_pred=model(data_train)
        loss = torch.sqrt(loss_func(y_pred,y_train))
        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())

        with torch.no_grad():
            y_pred = model(data_test)
            loss = torch.sqrt(loss_func(y_pred,y_test))
            test_loss.append(loss.item())
        if epoch%epochs_for_paint==0:
            print(f"e: {epoch}, train loss: {train_loss[-1]}, test loss: {test_loss[-1]}")
    plt.plot(np.arange(epoch+1),train_loss)
    plt.plot(np.arange(epoch+1),test_loss)
    plt.grid(True)
    plt.legend(['train','test'])
    plt.ylabel('RMSE')
    plt.xlabel('Epoch')
    plt.show()
```

```python
model = TORT()
optimizer = torch.optim.SGD(model.parameters(), lr=0.03)
```

```python
train(model,optimizer)
```

```python
with torch.no_grad():
    y_pred = model(data_test)
plt.plot(np.arange(len(y_pred)),y_pred)
plt.plot(np.arange(len(y_test)),y_test)
plt.grid(True)
plt.legend(['pred','truth'])
plt.show()
```

```python
with torch.no_grad():
    y_pred = model(data_train)
plt.plot(np.arange(len(y_pred)),y_pred)
plt.plot(np.arange(len(y_train)),y_train)
plt.grid(True)
plt.legend(['pred','truth'])
plt.show()
```

```python
def train_without_test(model, optimizer, epochs = 800, epochs_for_paint = 50):
    train_loss=[]
    loss_func=nn.MSELoss(reduction='mean')
    for epoch in range(0,epochs):
        optimizer.zero_grad()
        y_pred=model(data_train)
        loss = torch.sqrt(loss_func(y_pred,y_train))
        loss.backward()
        optimizer.step()
        train_loss.append(loss.item())
        if epoch%epochs_for_paint==0:
            print(f"e: {epoch}, loss: {train_loss[-1]}")
    plt.plot(np.arange(epoch+1),train_loss)
    plt.grid(True)
    plt.legend(['train'])
    plt.ylabel('RMSE')
    plt.xlabel('Epoch')
    plt.show()
```

```python
class TORT1(nn.Module):
    def __init__(self, n_marks):
        super().__init__()
        self.fc = nn.Linear(in_features=52, out_features=7, bias=True)
        self.tanh = nn.Tanh()
        self.conv1d = nn.Conv1d(in_channels=1, out_channels = n_marks, kernel_size=7)
        
    def forward(self,x):
        x = self.fc(x)
        x = self.tanh(x)
        return self.conv1d(x)
```

```python
lst=[]
CaptureSites = [f'CaptureSite_{i}' for i in range(29)]
week_columns = [f'week_{i}' for i in range(1,52+1)]
cols = CaptureSites + week_columns
for y in [2011,2012,2013,2014,2015,2016,2017,2018]:
    n_weeks = max(data[data.Year==y]['Week_num'])
    if n_weeks==53:
        n_weeks=52
    for week in range(1,n_weeks+1):
        week_array = [0]*52
        row = []
        temp = data[(data.Week_num == week) & (data.Year == y)]
        for site in CaptureSites:
            row.append(temp[temp.CaptureSite == site].shape[0])
        week_array[week-1] = 1
        row = row+week_array
        lst.append(row)

data_w = pd.DataFrame(lst,columns=cols) 
```
